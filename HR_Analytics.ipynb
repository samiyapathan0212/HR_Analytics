```python
sns.countplot(data=df, x='Attrition')
plt.title("Employee Attrition Count")
plt.show()

df['Attrition'].value_counts(normalize=True)*100
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='Gender', hue='Attrition')
plt.title("Attrition by Gender")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df, x='Department', hue='Attrition')
plt.xticks(rotation=30)
plt.title("Attrition by Department")
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.countplot(data=df, x='JobRole', hue='Attrition')
plt.xticks(rotation=45)
plt.title("Attrition by Job Role")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='MaritalStatus', hue='Attrition')
plt.title("Attrition by Marital Status")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['MonthlyIncome'], kde=True)
plt.title("Monthly Income Distribution")
plt.show()
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[18,25,35,45,60], labels=["18-25","26-35","36-45","46-60"])

sns.countplot(data=df, x='AgeGroup', hue='Attrition')
plt.title("Attrition by Age Group")
plt.show()
```

```python
df['IncomeGroup'] = pd.cut(df['MonthlyIncome'],
                           bins=[0,3000,6000,10000,20000],
                           labels=["Low","Medium","High","Very High"])

sns.countplot(data=df, x='IncomeGroup', hue='Attrition')
plt.title("Attrition by Income Level")
plt.show()
```

```python
df['TenureGroup'] = pd.cut(df['YearsAtCompany'],
                           bins=[0,2,5,10,20,40],
                           labels=["0-2","3-5","6-10","11-20","20+"])

sns.countplot(data=df, x='TenureGroup', hue='Attrition')
plt.title("Attrition by Tenure Group")
plt.show()
```
